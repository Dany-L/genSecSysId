```python
import numpy as np
import matplotlib.pyplot as plt 
%matplotlib widget 
names = ['silverbox_no-reg' , 'silverbox_reg']
```

```python

for name in names:
    y_hat = np.load(f"../data/{name}/predictions.npy") 
    y = np.load(f"../data/{name}/targets.npy")

    plt.figure(figsize=(10, 5))
    plt.plot(y[0,:,0], label='y')
    plt.plot(y_hat[0,:,0], label='y_hat')
    plt.legend()
    plt.title(f'{name} Predictions vs Targets')
    plt.xlabel('Time Steps')
    plt.ylabel('Values')
    plt.grid()
    plt.show()
```
